```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/Netflix Userbase.csv')
```

**These 3 commands here (below, helping me target the dataframes head & info)helped me explore my Dataset, looking to see the statistics and metrics behind this dataset.**


```python
print(df.head())
print(df.info())
print(df.describe())
```

Used a Line Plot to show **User Growth** over the 2-3 year timespan given in the Dataset, it seems User Groweth peaked in 2022, and decreased to normal levels in 2023, 2023 & 2021 both seem to have similar user growth.

```python
plt.figure(figsize=(10, 6))
df.groupby('Join Year')['User ID'].count().plot(marker='o')
plt.xlabel('Year')
plt.ylabel('Number of Users')
plt.title('User Growth Over Years')
plt.show()
```

Analyzed and used commands to construct a Bar Plot to see User Count by Country, looking to see which country has the moist interactions with Netflix. The USA and Spain seem to top this list. (Spain does suprise me.)

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Country')
plt.xticks(rotation=90)
plt.xlabel('Country')
plt.ylabel('Number of Users')
plt.title('User Count by Country')
plt.show()
```

Last but not least, I used some commands to create a Pie Chart, looking to see what types of Subscription types users favored on the Netflix platform from the dataset used. Most users favor the Basic subscription plan. The other plans the Standard and Premium also have a good percentage of users, not being far from taking dominance.

```python
plt.figure(figsize=(8, 8))
df['Subscription Type'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Subscription Types Distribution')
plt.ylabel('')
plt.show()
```
